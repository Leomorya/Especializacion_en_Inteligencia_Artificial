## Respuestas a las preguntas 
a) ¿Que es un algoritmo de aprendizaje automático (machine learning)?


Un algoritmo de aprendizaje automático (machine learning) es un algoritmo que puede aprender y mejorar a partir de datos, sin estar explicitamente programado, esto se logra con base en optimizaciones o reducciones de funciones error.

b)  ¿Que tipos de algoritmos de aprendizaje automático (machine learning) existen?

**Aprendizaje supervisado**: En este tipo de aprendizaje, los datos utilizados para entrenar el algoritmo están etiquetados con la salida deseada. El objetivo es aprender una  función que mapee las entradas a las salidas correctas, existe basicamente dos clases regresión para predecir un numero real e.j el valor de una acción o clasificación para predecir una clase e.j si es binario 0 o 1. algunos algoritmos son:

* Regresión lineal

* Regresión logística
* Máquinas de vectores de soporte (SVM)
*  Árboles de decisión y bosques aleatorios
*  Redes neuronales artificiales

**Aprendizaje no supervisado**: Aquí los datos de entrenamiento no están etiquetados, y el algoritmo debe encontrar patrones o estructuras interesantes en los datos por sí mismo. Ejemplos incluyen:

* Agrupamiento (clustering), como K-Means y DBSCAN
* Reducción de dimensionalidad, como Análisis de Componentes Principales (PCA) y t-SNE
* Asociación, como el algoritmo Apriori para análisis de reglas de asociación

**Aprendizaje semi-supervisado**: Una combinación de supervisado y no supervisado, donde tienes datos etiquetados y no etiquetados. Puede ser útil cuando tienes muchos datos pero solo algunos están etiquetados.

**Aprendizaje por refuerzo**: El algoritmo aprende a través de la interacción con un entorno, tomando decisiones y recibiendo retroalimentación en forma de recompensas o castigos. Ejemplos incluyen:

* Q-Learning
* Algoritmos de políticas, como el algoritmo REINFORCE

**Aprendizaje profundo (Deep Learning)**: Utiliza redes neuronales artificiales con múltiples capas (profundas) para aprender representaciones de datos de alto nivel. Ejemplos incluyen:

* Redes neuronales convolucionales (CNN) para tareas de visión por computadora
* Redes neuronales recurrentes (RNN) para procesamiento de secuencias, como en el procesamiento del lenguaje natural
* Redes neuronales generativas (GAN) para generar datos nuevos y realistas

c) ¿El algoritmo se basa o utiliza algún concepto de álgebra lineal para realizar la tarea para la cual esta diseñado?

Muchos algoritmos de aprendizaje automático se basan en conceptos o estructuras del álgebra lineal para optimizar o calcular regiones en espacios vectoriales de grandes dimensiones, los cuales no se pueden representar en un plano. Algunos ejemplos son:

* **Regresión lineal**: Este algoritmo utiliza ecuaciones lineales para modelar la relación entre variables de entrada y salida. En la regresión lineal, se ajusta una línea recta a los datos para minimizar el error entre las predicciones del modelo y los valores reales.

* **Máquinas de vectores de soporte (SVM)**: Las SVM utilizan conceptos como hiperplanos y operaciones de álgebra lineal para encontrar el mejor separador entre clases en un espacio dimensional superior. Esto permite clasificar nuevos datos en categorías basadas en su posición en relación con este separador.

* **Redes neuronales artificiales**: Las redes neuronales utilizan matrices de pesos y sesgos que se actualizan durante el proceso de entrenamiento. Estas matrices se multiplican y se suman en múltiples capas de neuronas, lo que implica operaciones fundamentales de álgebra lineal como la multiplicación de matrices y el cálculo de productos escalares.

* **Reducción de dimensionalidad**: Algoritmos como el Análisis de Componentes Principales (PCA) utilizan álgebra lineal para transformar datos de alta dimensión en un espacio de menor dimensión mientras se conserva la mayor cantidad posible de información relevante.

d) ¿Cómo este algoritmo utiliza el/los conceptos de algebra lineal?.

Existen diversas técnicas del álgebra lineal que utilizan los algoritmos de machine learning, como la multiplicación de matrices, la resolución de ecuaciones lineales, la descomposición espectral (valores y vectores propios), entre otros. Algunos ejemplos son:

* **Multiplicación de matrices y vectores**:  En muchos algoritmos, como las redes neuronales, SVM y PCA, se realizan operaciones de multiplicación de matrices y vectores para calcular transformaciones de datos, ajustar pesos y sesgos, y realizar proyecciones en espacios de menor dimensión.

* **Optimización y solución de sistemas de ecuaciones**: En algoritmos de regresión lineal y SVM, se utilizan técnicas de optimización que implican la solución de sistemas de ecuaciones lineales o la minimización de funciones objetivo, lo que a menudo se puede expresar y resolver usando álgebra lineal.
* **Descomposición de matrices**: Algoritmos como PCA y SVD (Descomposición de Valores Singulares) descomponen matrices de datos en componentes más simples que capturan la estructura subyacente de los datos. Esta descomposición se basa en conceptos de álgebra lineal y ayuda a reducir la dimensionalidad de los datos mientras se conserva la mayor cantidad posible de información relevante.
* **Distancias y similitudes**: Para agrupar los datos en clusters, los algoritmos de clustering a menudo calculan distancias o similitudes entre los puntos de datos. Esto implica operaciones matemáticas, como el cálculo de la distancia euclidiana o la similitud del coseno, que están basadas en conceptos de álgebra lineal.
*  **Espacios vectoriales y geometría**: Los algoritmos de clasificación, como las SVM, tratan de encontrar el hiperplano que mejor separa las clases en un espacio vectorial de alta dimensión. Este concepto se basa en la geometría y el álgebra lineal para encontrar una solución óptima.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML,display

# Definir los parámetros de la línea que separa el espacio (por ejemplo, y = mx + b)
m = 1.5  # Pendiente de la línea
b = 2    # Intercepto de la línea
# Establecer la semilla
np.random.seed(42)
# Generar puntos de datos aleatorios en el rango [0, 10] x [0, 10]
num_puntos = 60
x = np.random.uniform(0, 10, num_puntos)
y = np.random.uniform(0, 10, num_puntos)

# creamos las clases

y_line = x*m+b # puntos sobre la linea 
#creamos una mascara boleana
y_positiva = y[y_line > y+1]
x_positiva = x[y_line > y+1]

y_negativa = y[y_line < y-1]
x_negativa = x[y_line < y-1]

# limites de la grafica
xmin = -3
xmax = 13
ymin = -6
ymax = 18

# Graficar los puntos de datos y la línea que separa las clases
plt.scatter(x_positiva, y_positiva,marker='s',label='clase_positiva 0',color = 'red',alpha=0.8)
plt.scatter(x_negativa, y_negativa,marker='^',label='clase_negativa 1',color = 'blue' ,alpha=0.8)
plt.plot(np.linspace(xmin, xmax, 100), m*np.linspace(xmin, xmax, 100) + b, color='black', linestyle='--')
plt.xlabel('X')
plt.ylabel('Y')
plt.fill_between(np.linspace(xmin, xmax, 100), ymin, m*np.linspace(xmin, xmax, 100) + b, color='red', alpha=0.2)  # Rellena el área por debajo de la línea 
plt.fill_between(np.linspace(xmin, xmax, 100), m*np.linspace(xmin, xmax, 100) + b,ymax, color='blue', alpha=0.2)  # Rellena el área por encima de la línea 
plt.title('Conjunto de datos perfectamente separable')
plt.legend() # muestra la leyenda
plt.grid(True) # muestra la grilla 
plt.xlim(xmin, xmax)  # Límites para el eje x
plt.ylim(ymin, ymax)  # Límites para el eje y
plt.show()
```

Ya teniendo el conjunto de datos se implementara el perceptron.

```python
# construimos el calculo de z = w1*x1+w2*x2+b y step_function(z) para saber como clasifico
# para esto creamos un vector para cada clase, unamos las clases y generemos las etiquetas 

vector_clase_positiva = list(zip(x_positiva,y_positiva))
etiquetas_positivas = np.zeros(len(vector_clase_positiva))
vector_clase_negativa =  list(zip(x_negativa,y_negativa))
etiquetas_negativa = np.ones(len(vector_clase_negativa))
#generamos un solo vector y un solo vector de etiqueta 

vector_clases = np.concatenate((vector_clase_positiva , vector_clase_negativa ))
etiquetas = np.concatenate((etiquetas_positivas, etiquetas_negativa)) 
```

```python

# construimos la función de activación, para el caso la escalon unitaria o paso

def step_function(x):
    return np.where(x >= 0, 1, 0)
# generamos las predicciones 
def predicciones(pesos,sesgo,caracteristicas):
    z = caracteristicas@pesos+sesgo
    #print('z=',z)
    prediccion = step_function(z)
    #print('y=',prediccion)
    return np.array(prediccion)

# actulizamos los pesos y el sesgos, usando las medias 
def actualizar(caracteristicas,etiquetas_, predicciones_, pesos, sesgo, tasa_apredisaje):
    (etiquetas_-predicciones_)
    error = etiquetas_- predicciones_
    numero_datos = etiquetas_.shape[0]
    delta_w = (tasa_apredisaje/numero_datos)*(np.transpose(caracteristicas))@error
    delta_b = (tasa_apredisaje/numero_datos)*np.sum(error)
    w_actual = pesos+delta_w
    b_actual = sesgo+delta_b
    
    #print(f"error {error}\n")
    #print(caracteristicas)
    #print(f'delta pesos {delta_w}')
    #print(f' delta sesgo {delta_b}')
    #print(w_actual)
    #print(b_actual)
    return w_actual, b_actual
```

La función anterior de actualizar los pesos utiliza el metodo matricial, dado por la siguiente ecuación:

$$
\Delta\mathbf{W}= \eta \frac{1}{N}\mathbf{X}^T(\mathbf{y-\hat{y}})
$$ Donde $\Delta\mathbf{W}$ es la matriz que actualisa los pesos asi $\mathbf{W_actual} =\mathbf{W}+ \Delta\mathbf{W}$,  $\mathbf{X}^T $ es la matriz de caracteristica transpuesto, $\mathbf{y}$ es el vector de etiquetas y  $\mathbf{\hat{y}}$ el vector de etiquetas predichas por el perceptrón y $\eta$ es la taza de aprendisaje. cabe notar que aqui no se actulizan los pesos con cada instancia sino con una especie de promedio, parecido a un bacth.
para el sesgo se tiene:
$$
\Delta b= \eta \frac{1}{N}\sum_{i=1}^{n} (y_i-\hat{yi})
$$ 
donde $b$ se actuliza  $ b_{actual} = b+ \Delta b$, $n$ es el numero de muestras que se consideran para actulizar los pesos y sesgos para este caso todos.

Cabe anotar que uno puede actulizar los pesos y los sesgos con cada entrada, para este caso $n$ seria uno, pero la formula para $W$ seria algo distinta.

Ahora si implementemos el entrenamiento:


```python
def entrenamiento(epocas, matriz_caracteristicas, vector_etiquetas,tasa_aprendisaje_):
    # inicializamos los pesos y el sesgo
    np.random.seed(42)
    b_sesgo = np.random.rand(1)
    w_pesos = np.random.rand(2)
    for epoca in range(epocas):
         y_predict = predicciones(w_pesos, b_sesgo, matriz_caracteristicas)
         pesos_actual, sesgo_actual = actualizar(matriz_caracteristicas,vector_etiquetas ,
                                               y_predict, w_pesos, b_sesgo, tasa_aprendisaje_)
         #print('predicciones,',y_predict)
         #print("no actual",w_pesos,b_sesgo)
         w_pesos = pesos_actual
         b_sesgo = sesgo_actual
         #print("actual",w_pesos,b_sesgo)
    return w_pesos, b_sesgo    
```

```python
def gererar_data(numero_puntos,pendiente,corte_eje_y,semilla):
    """Esta funcion genera un conjunto de datos en el plano perfectamente separable,
    parametros: numero de puntos, pendiente de una recta, el punto de corte y un numero que es la semilla aleatoria

    retorna: un conjunto de puntos de longitud menor o igual a numero_puntos de entrada
    """
    
    m_= pendiente 
    b_= corte_eje_y
    num_puntos_ = numero_puntos
    x = np.random.uniform(0, 10, num_puntos_)
    y = np.random.uniform(0, 10, num_puntos_)
    # creamos las clases
    
    y_line = x*m_+b_ # puntos sobre la linea 
    #creamos una mascara boleana
    y_positiva_ = y[y_line > y+1]
    x_positiva_ = x[y_line > y+1]
    y_negativa_ = y[y_line < y-1]
    x_negativa_ = x[y_line < y-1]

   
    clase_positiva = list(zip(x_positiva_,y_positiva_))
    etiquetas_positivas = np.zeros(len(clase_positiva))
    clase_negativa =  list(zip(x_negativa_,y_negativa_))
    etiquetas_negativas = np.ones(len(clase_negativa))
    
    #generamos un solo vector y un solo vector de etiqueta 
    matriz_caracteristicas = np.concatenate((clase_positiva , clase_negativa ))
    vector_etiquetas = np.concatenate((etiquetas_positivas, etiquetas_negativas)) 

    
    from random import shuffle
    # creamos una tupla dato, etiqueta
    datos_con_etiquetas = list(zip(matriz_caracteristicas, vector_etiquetas)) 
    # barajamos los datos 
    shuffle(datos_con_etiquetas)
    # Desempaquetamos los datos barajados y las etiquetas barajadas
    datos_barajados, etiquetas_barajadas = zip(*datos_con_etiquetas)
    #separamos en x y y para que se vea mejor la data
    datos_barajados = np.array(datos_barajados)
    x_data =  datos_barajados[:,0]
    y_data =  datos_barajados[:,1]
    # creamos un diccionario con todo 
    data_label = {'x' : x_data,'y': y_data , 'target': etiquetas_barajadas}
    
    return data_label
```

Ahora solo nos falta una función que grafique el perceptron en cada epoca, es decir la linea que divide al conjunto y clasifica, para ver su apredisaje automatico y generar una animación.


```python
def grafico_perceptron(dataframe_datos,pesos,sesgo):


    x_max, y_max = df.max().values[:-1]+3
    x_min, y_min = df.min().values[:-1]-3
    
    # linea de separacion entre clases
    x_line = np.linspace(x_min, x_max, 100)
    y_line = -pesos[0]/pesos[1]*x_line- sesgo/pesos[1]
    
    
    
    # grafico de clases 
    ax = sns.scatterplot(data = dataframe_datos, x='x',y='y',hue='target',style='target', markers=['d','^'], palette='Set1',
                     s=20,alpha=0.8)
    
    
    linea, = plt.plot(x_line,y_line,linestyle='--', color = 'black', linewidth=1,label='Frontera entre clases', animated = True)
    
    relleno_0 = plt.fill_between(x_line, y_min, y_line, color='red', alpha=0.15,label ='perceptrón clasifica como clase : 0', animated = True)  # Rellena el área por debajo de la línea roja
    relleno_1 = plt.fill_between(x_line,y_line,y_max, color='blue', alpha=0.15,label ='perceptrón clasifica como clase : 1',animated = True)  # Rellena el área por debajo de la línea roja
    
    handles, labels = ax.get_legend_handles_labels()
    labels = ["Clase positiva : 0", "Clase negativa : 0"]
    ax.legend(handles=handles, labels=labels)
    
    plt.xlim(x_min, x_max)  # Límites para el eje x
    plt.ylim(y_min, y_max)  # Límites para el eje y
    
    # personalizar bien la layenda 
    #handles, labels = plt.gca().get_legend_handles_labels()
    #labels_personalizados = ["Clase positiva : 0", "Clase negativa : 1"]+labels[2:]
    
    #plt.legend(handles = handles,labels = labels_personalizados,loc=(1.1, 0.7) )
    plt.legend(labels=['0','2','2','2','1'])
    plt.title('Clasificación del perceptrón')
    
    
    figure = plt.gcf()  # Obtiene la figura actual
    
    
    return figure, linea, relleno_0, relleno_1
```

Finalmente se realiza la animación del entrenamiento del perceptrón 

```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML

tasa_de_aprendisaje = 0.2
contenedor = []

epocas = 120
for epoca in range(epocas):
    pesos_obt, sesgo_obt = entrenamiento(epoca, matriz_caract, etiquet, tasa_de_aprendisaje)
    fig, l,r1,r2 = grafico_perceptron(df, pesos_obt, sesgo_obt)  # obtien los artist, que son las grafican que se mueven en el tiempo 
     # Eliminar la leyenda actual
    handles, labels = fig.gca().get_legend_handles_labels()
    fig.gca().legend().remove()
    labels[-4:-3] = ["Clase positiva : 0", "Clase negativa : 1"]
    fig.gca().legend(handles[-5:], labels[-5:],loc=(1.05, 0.7))
    ax = plt.gca()

    # Crear el texto del título
    title = ax.text(0.7, 0.9, f"epoca={epoca}",
                    transform=ax.transAxes, fontdict={'size': 14,
                                                       "color": "purple",
                                                       "fontweight": "bold",
                                                       "fontstyle": 'italic'})
    contenedor.append([l,r1,r2,title])  # Añadir solo el objeto de ejes al contenedor

```

```python
animacion_perceptron = animation.ArtistAnimation(fig, contenedor, interval=100, blit=True)
fig.set_size_inches(10, 5) 
fig.subplots_adjust(left=0.1, right=0.6, bottom=0.1, top=0.9)
HTML(animacion_perceptron.to_jshtml())
```
